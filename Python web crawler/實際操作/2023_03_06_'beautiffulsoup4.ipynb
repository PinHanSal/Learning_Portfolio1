```python
from bs4 import BeautifulSoup
import requests, json, codecs

r1=requests.get('https://www.mol.gov.tw/1607/1632/1633/lpsimplelist',params={
    'Page':'1',
    'PageSize':'10'
})
b1=BeautifulSoup(r1.text,'html.parser')
b2= b1.find_all('h3')
for i in range(0, len(b2),1):
  # print(b2[i].find('a').attrs['title'])
  # print(b2[i].find('a').text)
  # print(b2[i].parent.find('div',{'class':'data'}).find_all('span')[2].text)
  # print('='*50)
  r2=requests.get('https://www.mol.gov.tw'+b2[i].find('a').attrs['href'])
  b3=BeautifulSoup(r2.text,'html.parser')
  with codecs.open('/content/news/data.txt','a','utf-8') as f:
    f.write(b2[i].parent.find('div',{'class':'data'}).find_all('span')[1].text+'\r\n')
    f.write(b2[i].find('a').text+'\r\n')
    f.write('='*50)
    f.write(b3.find_all('body')[1].text)
    f.write('='*50)

```

```python
from bs4 import BeautifulSoup
import requests, json, codecs
for p in range(1,6,1):
  r1 = requests.get('https://udn.com/api/more', params={
      'page':p,
      'id':'',
      'channelId':'1',
      'cate_id':'0',
      'type':'breaknews',
      'totalRecNo':'26301'
  })
      
  d=json.loads(r1.text)
  for i in range(0, len(d['lists']),1):
    print(d['lists'][i]['time']['date'],d['lists'][i]['title'])
    # print(b2[i].find('a').text)
    # print(b2[i].parent.find('div',{'class':'data'}).find_all('span')[2].text)
    # print('='*50)
    r2=requests.get('https://udn.com'+d['lists'][i]['titleLink'])
    if r2.status_code==200:
      b1=BeautifulSoup(r2.text,'html.parser')
    with codecs.open('/content/news/udn.txt','a','utf-8') as f:
      f.write(d['lists'][i]['time']['date']+'\r\n')
      f.write(d['lists'][i]['title']+'\r\n')
      section=b1.find('section',{'class':'article-content_editor'})
      if section!=None:
        for data in section.find_all('p'):
          f.write(data.text)
        f.write('='*50+'\r\n\r\n')



```

```python
import requests
from bs4 import BeautifulSoup

# 設定請求標頭
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36 Edg/110.0.1587.63'
}

# 發送請求，取得網頁內容
res = requests.get('https://soa.tainan.gov.tw/Api/Service/Get/4a941a39-bc95-43a9-890c-946facf45cf9', headers=headers)
html = res.text

# 使用BeautifulSoup解析HTML文件
soup = BeautifulSoup(html, 'html.parser')

# 找到所有縣市的行
city_rows = soup.find_all('tr', class_='countyTr')

# 逐一取得每個縣市的溫度
for row in city_rows:
    # 取得縣市名稱
    city_name = row.find('a').text.strip()
    
    # 取得縣市溫度
    city_temp = row.find('td', class_='temp').text.strip()
    
    # 印出縣市名稱與溫度
    print(city_name, city_temp)
```
