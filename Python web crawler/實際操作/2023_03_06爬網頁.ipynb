```python
import requests,json
for p in range(1,10):
  r1 = requests.get('https://udn.com/api/more?page=4&id=&channelId=1&cate_id=0&type=breaknews&totalRecNo=26320', params={
      'page':p,
      'id':'',
      'channelId':'1',
      'cate_id':'0',
      'type':'breaknews',
      'totalRecNo':'26320',
      
  })
  d = json.loads(r1.text)

  for i in range(0,len(d['lists'],),1):
    print(d['lists'][i]['time']['date'],d['lists'][i]['title'])
```

```python
import requests, json

keyword = input('關鍵字:')
page = input('頁碼:')
for p in range(1, int(page)):
    r1 = requests.get('https://xapi.pcone.com.tw/search/commodities',
                      headers={
                          'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36 Edg/110.0.1587.63'

                      },
                      params={
                          'site': 'pcone',
                          'direction': 'asc',
                          'page': p,
                          ' q': keyword,
                          'sorting': 'default',
                          'device': 'desktop',
                          'aid': 'null'
                      })
    d = json.loads(r1.text)

    for i in range(0, len(d['data']['commodities'], ), 1):
        print(d['data']['commodities'][i]['name'], d['data']['commodities'][i]['price'])

```
