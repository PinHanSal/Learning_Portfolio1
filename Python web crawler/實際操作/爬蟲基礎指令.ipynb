```python
import requests
```

```python
url = 'http://teaching.bo-yuan.net/test/requests/'
params = {'action':'get'}
data = {'id':'30', 'name':'陳品翰', 'address':'32faf0'}
```

```python
print(requests.get(url, params=params).text)
print(requests.delete(url, params=params, data=data).text)
print(requests.put(url, params=params, data=data).text)
print(requests.patch(url, params=params, data=data).text)
print(requests.post(url, params=params, data=data).text)
```
